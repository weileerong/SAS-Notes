```sas
data job;
infile 'job.txt';
input Y X1 X2 X3 X4;
run;
```

```sas
proc print data=job(obs=5);
run;
```

```sas
*scatterplot;
proc sgscatter data=job;
matrix y x1 x2 x3 x4 / diagonal=(histogram);
run;
```

```sas
proc corr data=job nosimple;
var y x1 x2 x3 x4;
run;
```

```sas
proc reg data=job;
model y = x1 x2 x3 x4;
run;
```

```sas
proc reg data=job;
model y = x1 x2 x3 x4 / selection=ADJRSQ;
run;
```

```sas
proc reg data=job;
model y = x1 x2 x3 x4 / selection=stepwise;
run;
```
