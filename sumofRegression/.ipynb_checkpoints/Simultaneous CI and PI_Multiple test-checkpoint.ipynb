```sas
* Bonferrroni simultaneous ci for points on the regression curve;
proc glm data=pg alpha=.01666666; *calculated alpha;
model steam = glycerin / CLPARM;
estimate 'At .6 lbs' intercept 1 glycerin  .6;
estimate 'At .7 lbs' intercept 1 glycerin  .7;
estimate 'At .8 lbs' intercept 1 glycerin  .8;
run;
```

```sas
/****************************************************************************************/
/*
/*Working-Hotelling simultaneous ci for points on the regression curve
/*
/****************************************************************************************/

data extra;
input steam glycerin;
datalines;
. 0.6
. 0.7
. 0.8
;
run;

data extended;
  set pg extra;
run;

proc print data = extended; run;

proc reg data=extended outest=ests;
  model steam=glycerin;
  output out=ExtendedFit p=HatSteam stdi=s;
run; quit;

proc print data = ExtendedFit; run;

data CompWH;
  set ExtendedFit;
  fvalue=finv(0.95,2,23);
  W=sqrt(2*fvalue);
  LowWH= HatSteam - W*s;
  HighWH = HatSteam + W*s;
  if steam ne . then delete;
  keep glycerin HatSteam LowWH HighWH;
run;

proc print data=CompWH;
run;
```

```sas
/****************************************************************************************/
/*
/* Bonferrroni and Scheffe simultaneous ci for prediction
/*
/****************************************************************************************/
proc reg data=extended;
  model steam=glycerin;
  output out=ExtPred p=HatY stdp=sp;
run;
data PredInters;
  set ExtPred;
  /**/
  fvalue=finv(0.95,3,23);
  W=sqrt(3*fvalue);
  LowSCH= HatY - W*sp;
  HighSCH = HatY + W*sp;
  /**/
  tvalue=tinv(0.99166666,23);
  LowBF = HatY - tvalue*sp;
  HighBF = HatY+tvalue*sp;
  if steam ne . then delete;
  keep glycerin HatY LowSCH HighSCH LowBF HighBF;
run;
proc print data=PredInters;
run;
```

```sas
*Multiple test;
data p_values;
input raw_p;
datalines;
0.001
0.002
0.004
0.005
0.008
0.011
0.032
0.045
0.062
0.112
;
run;
proc multtest pdata = p_values fdr bonferroni;
run;
```

```sas
data NewPred;
input steam temp frezday;
datalines;
. 35 0
. 35 7
. 35 14
. 35 28
;

data extended;
 set pg;
 keep steam temp frezday;
run;
data extended;
 set extended NewPred;
run;
proc print data = extended;
run;


proc reg data=extended;
 model steam = temp frezday;
 output out=ExtendedFit p=SteamHat STDI=sp STDP=s;
run; quit;

proc print data=ExtendedFit; run;

data intervals;
 set ExtendedFit;
 W = sqrt( 3*finv(.95,3,22));
 She = sqrt( 4*finv(.95,4,22));
 bonlev=1 - (.05/8);
 Bon = tinv(bonlev,22);
 mean_bon_L = SteamHat - Bon*s;
 mean_bon_H = SteamHat + Bon*s;
 mean_WH_L = SteamHat - W*s;
 mean_WH_H = SteamHat + W*s;
 pred_Sch_L = SteamHat - She*sp;
 pred_Sch_H = SteamHat + She*sp;
 pred_bon_L = SteamHat - Bon*sp;
 pred_bon_H = SteamHat + Bon*sp;
 if steam ne . then delete;
run;
proc print data=intervals;
 var frezday mean_bon_L mean_bon_H 
 mean_WH_L mean_WH_H pred_Sch_L 
 pred_Sch_H pred_bon_L pred_bon_H;
run;
```
