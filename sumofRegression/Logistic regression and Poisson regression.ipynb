```sas
proc genmod data=crab descending;
model y= width/link = logit dist = binomial;
run;
```

```sas
proc genmod data=crab descending; /* Changes the ordering*/
model y=width / link=logit dist=binomial type3 covb ; /*type3=LRT based ANOVA table,
														covb = output covariance matrix */
contrast 'LRT test'  width 1 ; /*LRT */
contrast 'Wald test' width 1 / wald; /*Wald*/
estimate 'effect of width' width 1/exp;
estimate 'log odds at 21cm' intercept 1 width 21 /exp; /*Estimates of a Crab of width 21cm with wald test*/
run;
```

```sas
/****************************************************************************

* confidence intervals ;

/****************************************************************************/

proc genmod data=crab descending; /* Changes the ordering*/
model y=width / link=logit dist=binomial lrci waldci; /*default is wald intervals, 
														lrci produes lrci only,
														lrci + waldci gives both*/
run;
```

```sas
/****************************************************************************
*
*
*	Additive model for the crab data with width and color (class variable)
*
*
****************************************************************************/

proc genmod data=crab descending;
class color / param=glm;
model y = width color / link=logit dist=binomial lrci waldci type3;
estimate 'color=1 vs. color=2' color 1 -1 0 0 / exp;
estimate 'color=2 vs. color=4' color 0 1 0 -1/ exp;
estimate 'effect of width' width 1/exp; 
run;
```

```sas
/****************************************************************************
*
*
*	Model for the crab data with interaction
*
*
****************************************************************************/

proc genmod data=crab descending;
class color / param=glm;
model y = width color width*color / link=logit dist=binomial lrci waldci type3;
estimate 'effect of width in color 3' width 1 width*color 0 0 1 0 /exp; 
estimate 'effect of width in color 4' width 1 width*color 0 0 0 1/exp;
run;
```
